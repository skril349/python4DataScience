```python
import numpy as np
```

```python
ruta = "medallas.csv"
array = np.genfromtxt(ruta)
array
```

```python
array = np.genfromtxt(ruta, delimiter=",")
array
```

```python
array = np.genfromtxt(ruta, delimiter=",",filling_values=0)
array
```

```python
array = np.genfromtxt(ruta, delimiter=",",filling_values=0,skip_header=True )
array
```

```python
array = np.genfromtxt(ruta, delimiter=",",filling_values=0,skip_header=True, dtype= int )
array
```

```python
array_ejemplo = np.array([[1,2,3],
                         [4,5,6]])
```
