```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
ruta = "datos_tarjetas/tarjetas_credito.csv"
df = pd.read_csv(ruta,header=None, names=['A'])
dfn = df['A'].str.split(',', expand=True)
 
dfn.to_csv('tarjetas2.csv', index=False, header=False)
 
dfn.head()
 
dft = pd.read_csv('tarjetas2.csv')
dft.head()
 
# cambiamos el formato de las columnas a str (sino tira error)
dft[:0].astype(str)
 
# y eliminamos las comillas de las columnas
dft.columns = dft.columns.str.strip('"')
 
# tambien en la columna clase
dft['Clase'] = dft['Clase'].str.strip('"').astype(int)
 
dft.head()
```

```python
escala = MinMaxScaler(feature_range=(0,1))
```

```python
normalizado = escala.fit_transform(dft)
df_normalizado = pd.DataFrame(data = normalizado, columns = dft.columns)
df_normalizado
```

```python
X = df_normalizado.drop("Clase", axis = 1)
```

```python
y = df_normalizado["Clase"]
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
random_forest = RandomForestClassifier()
```

```python
random_forest.fit(X_entrena, y_entrena)
```

```python
random_forest.score(X_prueba, y_prueba)
```
