```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
df =pd.DataFrame()
df["Area"] = [2600,3000,3200,3600,4000]
df["Precio"] = [550000,565000,610000,680000,725000]
```

```python
df
```

```python
plt.scatter(df["Area"],
           df["Precio"]);
```

```python
X = df[["Area"]]
```

```python
X
#Es un dataframe per posar els [[]]
```

```python
y = df["Precio"]
```

```python
modelo = linear_model.LinearRegression()
```

```python
modelo.fit(X,y)
```

```python
modelo.predict([[3300]])
```

```python
plt.scatter(df["Area"],
           df["Precio"]);
plt.plot(df["Area"],modelo.predict(X))
```

Crea un modelo de regresión lineal para predecir el salario basándose en la edad.

Asegurate de mantener la información del gráfico anterior:



Titulo del gráfico: 'Relación entre Edad y Salario'

Etiqueta del eje x: 'Edad'

Etiqueta del eje y: 'Salario'

Activa la rejilla en el gráfico (grid)

```python
data = {'Edad': [25, 30, 35, 40, 45],
        'Salario': [3000, 3500, 4000, 4500, 5000]}
df = pd.DataFrame(data)
plt.scatter(df["Edad"],df["Salario"])
plt.grid(True)
plt.title("Relación entre Edad y Salario")
plt.xlabel("Edad")
plt.ylabel("Salario")

X = df[["Edad"]]
y = df["Salario"]
modelo = linear_model.LinearRegression()
modelo.fit(X,y)
plt.plot(df["Edad"],modelo.predict(X));
```
