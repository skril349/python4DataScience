```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Crear el DataFrame con los datos
df = pd.DataFrame({
    "frutas": ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos'],
    "cantidades": [40, 25, 15, 20]
})
 
# Explode para separar todas las porciones
explode = (0.1, 0.1, 0.1, 0.1)  # explota todas las frutas
 
# Crear el grfico de pastel 
plt.pie(df['cantidades'], labels=df['frutas'], autopct='%1.1f%%', explode=explode)
plt.title('Distribucin de frutas en el almacn')
 
# Usar el estilo 'dark_background'
plt.style.use('dark_background')
```
