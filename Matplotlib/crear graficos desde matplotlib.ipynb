```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ruta = "Iluminacion.csv"
df = pd.read_csv(ruta)
```

```python
df
```

```python
help(plt.hexbin)
```

```python
fig, ax = plt.subplots()
plt.hexbin(df["Latitud"],df["Longitud"],C = df["Luminosidad"],gridsize=50,cmap="Greens")
ax.set_facecolor("#000000")
plt.colorbar(label="Nivel de sombras")
plt.xlabel("Latitud")
plt.ylabel("Longitud")
plt.title("Luminosidad en Villa Luz");
```
