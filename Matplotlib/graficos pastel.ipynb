```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame({
    "etiquetas":["Manzanas","Bananas","Naranjas","Uvas"],
    "Cantidad":[30,60,80,20]
})
```

```python
df
```

```python
plt.pie(df["Cantidad"]);
```

```python
plt.pie(df["Cantidad"],labels=df["etiquetas"],autopct="%1.1f%%");
```

```python
plt.pie(df["Cantidad"],labels=df["etiquetas"],autopct="%1.1f%%",colors = ["#87CEEB","#6495ED","#000080","#40E0D0"]);
```

```python
plt.pie(df["Cantidad"],labels=df["etiquetas"],autopct="%1.1f%%",colors = ["#87CEEB","#6495ED","#000080","#40E0D0"], 
        explode=(0,0,0,0.2))
plt.title("Mi Pie Chart");
```
