```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.grid()
```

```python
plt.plot(x,y,'o')
plt.grid()
```

```python
aleatorio = np.random.RandomState(0)
marcadores = ['o','x','+','v','s','d']
```

```python
for marcador in marcadores:
    plt.plot(aleatorio.rand(5),aleatorio.rand(5),marcador)
```

```python
plt.plot(x,y,':ob')
```

```python
plt.plot(x,y,'-p',color="red",markersize = 10,linewidth=5, markerfacecolor="white")
```

```python
plt.scatter(x,y,marker="v")
```

```python
plt.scatter(aleatorio.randn(100),aleatorio.randn(100), c = aleatorio.randn(100))
```

```python
plt.scatter(aleatorio.randn(100),aleatorio.randn(100), c = aleatorio.randn(100), s= aleatorio.randn(100)*1000, alpha=0.3, cmap='viridis');
plt.colorbar()
```

```python
# Datos para las series del grfico
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

plt.scatter(x1,y1,marker="o",c="blue", label="Serie 1")
plt.scatter(x2,y2,marker="^",c="green", label = "Serie 2")
plt.legend()
```
