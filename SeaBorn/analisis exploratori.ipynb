# Crear visualitzacions

Relacions entre variables numèriques

```python
#Configurem l'estil
sns.set(style="whitegrid")

#Creant el gràfic de relació amb relplot

g = sns.relplot(data = diamantes,
               x="carat",
               y="price",
               kind = "scatter",
               height=6,
               aspect=1.5)

#Posem titol
g.fig.suptitle("Relació entre el preu i el pes en quilats dels diamants",
              va = 'baseline',
              ha = "center");
g.set_axis_labels("Pes en Quilats", "Preu");
```

Podem observar una variabilitat de preus i una relació positiva, doncs si augmenta el pes, augmenta el preu.
Hi ha altres factors que fan variar el preu, doncs veiem molta variabilitat.
Veiem que hi ha bastants outliers ("diamants amb característiques úniques")

## Distribucións de variables numèriques

```python
#Creem un histograma
g = sns.displot(data=diamantes,
               x="price",
               kind="hist",
               kde = True,
               color="blue",
               height=4,
               aspect=2)
g.fig.suptitle("Distribució dels preus dels diamants",
              va = "baseline",
              ha="center")
g.set_axis_labels("Preu ($)", "Freqüència");
```

```python
#Creem la visualització de la distribució pel pes en quilats
g = sns.histplot(data=diamantes,
               x="carat",
               kde = True,
               color="green",
               )
g.set_title("Distribució dels Pes en Quilats")
g.set(xlabel="Carat",ylabel= "Freqüència");
```

## Comparacio de variables numèriques per catgories

```python
g = sns.catplot(data= diamantes,
               kind = "box",
               x="cut",
               y="price",
               height=5,
               aspect=2,
               order=["Fair","Good","Very Good", "Premium","Ideal"])
g.fig.suptitle("Comparació dels preus dels diamants per qualitat del tall",
              va = "baseline",
              ha="center")
g.set_axis_labels("Qualitat del tall","Preu ($)");
```

```python
#Utilitxzant joinplot per visualitzar la relació entre depth i price amb un gràfic de densitat kernel

g = sns.jointplot(data = diamantes,
               x = "depth",
               y = "price",
               kind = "kde",
               fill = True,
               space = 0,
               color = "purple")
# Cambiem etiquetes
g.set_axis_labels("Profunditat del diamant (%)","Preu ($)");
g.fig.suptitle("Relació entre la profunditat i el preu dels diamants",
              va = "baseline",
              ha="center",
              fontsize = 16);
```

```python
g = sns.pairplot(data=diamantes,
                corner = True)
g.fig.suptitle("Relació entre variables numèriques del dataset diamants",
              va = "baseline",
              ha="center",
              fontsize = 16);
```
