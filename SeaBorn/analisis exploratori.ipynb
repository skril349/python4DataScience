# Crear visualitzacions

Relacions entre variables num√®riques

```python
#Configurem l'estil
sns.set(style="whitegrid")

#Creant el gr√†fic de relaci√≥ amb relplot

g = sns.relplot(data = diamantes,
               x="carat",
               y="price",
               kind = "scatter",
               height=6,
               aspect=1.5)

#Posem titol
g.fig.suptitle("Relaci√≥ entre el preu i el pes en quilats dels diamants",
              va = 'baseline',
              ha = "center");
g.set_axis_labels("Pes en Quilats", "Preu");
```

Podem observar una variabilitat de preus i una relaci√≥ positiva, doncs si augmenta el pes, augmenta el preu.
Hi ha altres factors que fan variar el preu, doncs veiem molta variabilitat.
Veiem que hi ha bastants outliers ("diamants amb caracter√≠stiques √∫niques")

## Distribuci√≥ns de variables num√®riques

```python
#Creem un histograma
g = sns.displot(data=diamantes,
               x="price",
               kind="hist",
               kde = True,
               color="blue",
               height=4,
               aspect=2)
g.fig.suptitle("Distribuci√≥ dels preus dels diamants",
              va = "baseline",
              ha="center")
g.set_axis_labels("Preu ($)", "Freq√º√®ncia");
```

```python
#Creem la visualitzaci√≥ de la distribuci√≥ pel pes en quilats
g = sns.histplot(data=diamantes,
               x="carat",
               kde = True,
               color="green",
               )
g.set_title("Distribuci√≥ dels Pes en Quilats")
g.set(xlabel="Carat",ylabel= "Freq√º√®ncia");
```

## Comparacio de variables num√®riques per catgories

```python
g = sns.catplot(data= diamantes,
               kind = "box",
               x="cut",
               y="price",
               height=5,
               aspect=2,
               order=["Fair","Good","Very Good", "Premium","Ideal"])
g.fig.suptitle("Comparaci√≥ dels preus dels diamants per qualitat del tall",
              va = "baseline",
              ha="center")
g.set_axis_labels("Qualitat del tall","Preu ($)");
```

```python
#Utilitxzant joinplot per visualitzar la relaci√≥ entre depth i price amb un gr√†fic de densitat kernel

g = sns.jointplot(data = diamantes,
               x = "depth",
               y = "price",
               kind = "kde",
               fill = True,
               space = 0,
               color = "purple")
# Cambiem etiquetes
g.set_axis_labels("Profunditat del diamant (%)","Preu ($)");
g.fig.suptitle("Relaci√≥ entre la profunditat i el preu dels diamants",
              va = "baseline",
              ha="center",
              fontsize = 16);
```

```python
g = sns.pairplot(data=diamantes,
                corner = True)
g.fig.suptitle("Relaci√≥ entre variables num√®riques del dataset diamants",
              va = "baseline",
              ha="center",
              fontsize = 16);
```
