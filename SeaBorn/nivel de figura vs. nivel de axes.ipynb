```python
import seaborn as sns
```

```python
pinguinos = sns.load_dataset("penguins")
```

```python
sns.displot(data = pinguinos,
           x = "flipper_length_mm",
           hue = "species",
           multiple = "stack");
```

```python
sns.histplot(data = pinguinos,
           x = "flipper_length_mm",
           hue = "species",
           multiple = "stack");
```

```python
sns.displot(data = pinguinos,
           x = "flipper_length_mm",
           hue = "species",
           col="species");
```

```python
sns.relplot(data = pinguinos,
           x="flipper_length_mm",
           y = "bill_length_mm",
           col = "sex");
```

```python
graph = sns.relplot(data = pinguinos,
           x="flipper_length_mm",
           y = "bill_length_mm",
           col = "sex")
graph.set_axis_labels("Largo Aletas (mm)", "Largo del pico (mm)");
```
