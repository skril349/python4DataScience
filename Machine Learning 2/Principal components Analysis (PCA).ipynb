```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
iris
```

```python
X = iris.data
y = iris.target
```

```python
X
```

```python
y
```

```python
# Estandarització de les dades
X_centrado = X- np.mean(X, axis=0)
```

```python
X_centrado
```

```python
pca = PCA(n_components=2)
```

```python
X_pca = pca.fit_transform(X_centrado)
```

```python
especies = ["setosa","versicolor", "virginica"]
```

```python
plt.figure(figsize=(8,6))
for i in range(0,3):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label = especies[i])
    
plt.xlabel("Primer compoennte principal")
plt.ylabel("Segundo componente principal")
plt.legend()
plt.title("PCA de Iris");
```
